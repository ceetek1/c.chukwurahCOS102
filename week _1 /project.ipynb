```python
# Simple interest 
def smpin(p,r,t):
    p*(1 + (r/100)*t)
    return p*(1 + (r/100)*t)
p = float(input("Enter the principle: "))
r = float(input("Enter the rate: "))
t = float(input("Enter the time: "))
print('Your amount is ',smpin(p,r,t))
```

```python
# Compound Interest 
def cmpin(p,r,n,t):
    return p*(1+r/n)/n*t

a = cmpin(3000,3,0.8,3)
print (a)
```

```python
# Annuity plan 
def annp(p,m,t,r,n):
    return p*m*t*((1+r/n)/n*t-1)/r/n
# p = float(input("Enter the principle: ")
    

a = annp(300000,300,5,6,0.2)
print(a)
```
