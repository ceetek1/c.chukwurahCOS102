```python
# Simple interest 
def smpin():
    p = float(input("Enter the principle: "))
    r = float(input("Enter the rate: "))
    t = float(input("Enter the time: "))
    a=p*(1 + (r/100)*(t))
    return a
print('Your amount is:',smpin())
```

```python
# Compound Interest 
def cmpin(p,r,n,t):
    return p*(1+r/(100*n))**(n*t)

a = cmpin(3000,3,8,5)
a
```

```python
# Annuity plan 
def annp(p,m,t,r,n):
    return p*m*t*((1+r/(100*n))**(n*t)-1)/(r/n)
# p = float(input("Enter the principle: ")
# The rate is meant to be in percentage 

a = annp(300000,300,5,6,0.2)
a
```
