```python
import tkinter as tk

class Delivery:
    def __init__(self, location, weight):
        self.location = location.lower()
        self.weight = weight

    def calculate_fee(self):
        if self.location == "pau":
            if self.weight >= 10:
                return 2000
            else:
                return 1500

        elif self.location == "epe":
            if self.weight >= 10:
                return 5000
            else:
                return 4000

        else:
            return None


# GUI Function
def calculate():
    location = location_entry.get()
    
    try:
        weight = float(weight_entry.get())
    except:
        result_label.config(text=" Please enter a valid weight")
        return

    delivery = Delivery(location, weight)
    fee = delivery.calculate_fee()

    if fee is None:
        result_label.config(text=" Invalid location! Use PAU or Epe")
    else:
        result_label.config(text=f" Delivery Fee: ₦{fee}")


# GUI Setup
root = tk.Tk()
root.title("Delivery Fee Calculator")
root.geometry("400x300")

# Widgets
tk.Label(root, text="Enter Location (PAU or Epe):").pack(pady=10)
location_entry = tk.Entry(root)
location_entry.pack()

tk.Label(root, text="Enter Package Weight (kg):").pack(pady=10)
weight_entry = tk.Entry(root)
weight_entry.pack()

tk.Button(root, text="Calculate Fee", command=calculate).pack(pady=15)

result_label = tk.Label(root, text="")
result_label.pack(pady=20)

# Run app
root.mainloop()
```
