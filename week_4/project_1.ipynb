```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
airb_df = pd.read_csv('NYC Airbnb.csv')
airb_df
```

```python
airb_df.isna().sum()
```

```python
airb_df.dropna(inplace=True)
airb_df.isna().sum()
```

```python
pivot = airb_df.pivot_table(index='neighbourhood_group', columns='room_type', values='id', aggfunc='count')
plt.figure(figsize=(10,8))
sns.heatmap(pivot, annot=True, fmt='g', cmap='coolwarm')
plt.title('Listings count heatmap by borough and room type')
plt.show()
```

In the data above it shows that Manhattan has many entire home/apt,
Brooklyn has many private room type 
manhattan has many shared rooms
in conclusion manhattan has the most apartments
