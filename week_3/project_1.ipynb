```python
# GUI Iterface for logistics
import tkinter as tk 
from tkinter import messagebox as msgbox
import csv 


def open_employees():
    employees = []
    with open('GIG-logistics.csv',newline='') as file:
        reader = csv.DictReader(file)
        for row in reader:
           employees.append(row)
        return employees
    
def search_employee():
    name = name_entry.get().strip().lower()
    dept = dept_entry.get().strip().lower()

    found = False 
    department_members = []               

    for emp in employees:# loops through the whole csv file
        emp_name = emp["Name"].strip().lower()
        emp_dept = emp["Department"].strip().lower()

        if emp_name == name and emp_dept == dept:
            found= True
        
        if emp_dept == dept:
            department_members.append(emp["Name"])
        
    if found:
        msg = f"WELCOME {name.title()} \n Other members in the{dept.title()}:\n"
        msg+= "\n".join(department_members)
        msgbox.showinfo("GREAT",msg)
    else:
        msgbox.showerror("Error","SORRY EMPLOYEE NOT FOUND")

        
employees = open_employees()

        
window = tk.Tk()
window.title("GIG-Logistics")
window.geometry("400x550")
window.config(background="#03fca9")
# Labels
tk.Label(window, text="Enter Name").pack()
name_entry=tk.Entry(window)
name_entry.pack()

tk.Label(window, text="Enter Department").pack()
dept_entry= tk.Entry(window)
dept_entry.pack()

tk.Button(window, text="CONFIRM", command=search_employee).pack(pady=10)

window.mainloop()



```
