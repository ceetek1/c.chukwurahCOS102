```python
# GUI Iterface for logistics
import tkinter as tk 
from tkinter import messagebox as msgbox
from tkinter import font
import csv 


def open_employees():
    employees = []
    with open('GIG-logistics.csv',newline='') as file:
        reader = csv.DictReader(file)
        for row in reader:
           employees.append(row)
        return employees
    
def search_employees():
    name = name_entry.get().strip().lower()
    dept = dept_entry.get().strip().lower()

    found = False 
    department_members = []

    for emp in employees:
        emp_name = emp["Name"].strip().lower()
        emp_dept = emp["Department"].strip().lower()

        if emp_name == name and emp_dept == dept:
            found= True
        
        if emp_dept == dept:
            department_members.append(emp["NAME"])
        
        if found:
            msg = f"WELCOME{name.title()} \n Other members in the{dept.title()}:\n"
            msg+= "\n".join(department_members)
            msgbox.showinfo("GREAT",msg)
        else:
            msgbox.showerror("Error","SORRY EMPLOYEE NOT FOUND")

        
        employees = open_employees()
root = tk.Tk()
root.title("GIG-Logistics")
root.geometry("400x250")
gig_font = font.Font(family="Helvetica", size=28, weight="bold", slant= "italic")

# Labels
tk.Label(root, text="Enter Name").pack()
name_entry=tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Enter Department")
dept_entry= tk.Entry(root)
dept_entry.pack()

tk.Button(root, text="CONFIRM", command=search_employees).pack(pady=10)

root.mainloop()



```
