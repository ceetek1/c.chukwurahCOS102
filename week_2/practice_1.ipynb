```python
tuple = ('Ekiti', 750, 'Oshogbo', 250, 'Akure', 500)
s_tuple = ('Abekuta', 300, 'Ogbomosho')

# Prints the complete tuple 
print(tuple)

# Prints the last element of the tuple 
print (tuple[-1])

# Prints elements of the tuple starting from 2nd tii 3rd 
print (tuple[2:4])

# Prints elements of the tuple starting from 3rd element 
print (tuple[3:])

# Prints the contents of the tuple twice 
print (s_tuple * 3)

# Prints concatenated tuples 
print (tuple + s_tuple)
```

```python
# Input from user
m = float(input("Enter mass in kilogram: "))

# Constant value for the speed of light in m/s
c = 299792458

# Calculating energy using Einstein's equqation 
energy = m * c **2 

# Displaying the result 
print (f"The energy equivalent to {m} kg of mass is {energy} joules.")
```
