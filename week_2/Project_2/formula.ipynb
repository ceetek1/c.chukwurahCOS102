```python
# A calculator that performs CUBIC,QUADRATIC AND QUARTIC equations 
print("------------------------WELCOME------------------------")
import numpy as np 
def quad ():
    print("Enter coefficients for the quadratic equation ax^2 + bx + c= 0")
    a = float(input("Enter input: ")) # Allows the users enter their value 
    b = float(input("Enter input: "))
    c = float(input("Enter input: "))
    coefficients = (a,b,c)
    roots = np.roots(coefficients)

    print ("Roots of the quadratic equation:")
    for root in roots:
        print(root)
    
def cubic ():
    print("Enter coefficients for cubic equation ax^3 + bx^2 + cx = 0")
    a = float(input("Enter input: "))
    b = float(input("Enter input: "))
    c = float(input("Enter input: "))
    d = float(input("Enter input: "))
    coefficients = (a,b,c,d)
    roots = np.roots(coefficients)

    print ("Roots of the cubic equation:")
    for root in roots:
        print(root)
    
def quartic ():
    print("Enter coefficients for the quartic equation ax^4 + bx^3 + cx^2 + dx + e = 0")
    a = float(input("Enter input: "))
    b = float(input("Enter input: "))
    c = float(input("Enter input: "))
    d = float(input("Enter input: "))
    e = float(input("Enter input: "))
    coefficients = (a,b,c,d,e)
    roots = np.roots(coefficients)

    print ("Roots of the quatric equation:")
    for root in roots:
        print(root)

while  True:

    operation = "CUBIC\nQUADRATIC\nQUARTIC"
    print (operation)
    user_input = input("Enter Your prefered operation: ").strip().upper()
    
    

    if user_input == "CUBIC":
       cubic()
    elif user_input == "QUADRATIC":
        quad()
    elif user_input == "QUARTIC":
        quartic()
    else:
        print ("INVALID INPUT, ENTER 'CUBIC', 'QUADRATIC', 'QUATRIC' ")
        continue



    quit_input =(input (f"Do You Want to continue YES OR NO")).strip().lower()
    if quit_input == "yes":
       continue
    elif quit_input == "no":
        print("GOOD BYE")
        break
    else:
        print("INVALID INPUT , CONTINUING...........")
        continue



```
